Crie um algoritmo em R que peça as 4 notas bimestrais e mostre a média.

```
nota1=as.numeric(readline("Digite a nota 1: "))
nota2=as.numeric(readline("Digite a nota 2: "))
nota3=as.numeric(readline("Digite a nota 3: "))
nota4=as.numeric(readline("Digite a nota 4: "))
print(sprintf("A média é %s", (nota1+nota2+nota3+nota4)/4))
```

Crie um algoritmo em R que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

```
lado=as.numeric(readline("Informe o valor do lado do quadrado: "))
print(sprintf("A área do quadrado é %.2f. O dobro da área do quadrado é %.2f", lado*lado, 2*lado*lado))
```

Crie um algoritmo em R que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.


```
valor_hora=as.numeric(readline("Quanto você ganha por hora? "))
quant_horas=as.numeric(readline("Quantos horas você trabalhou no mês? "))
print(sprintf("Seu salário do mês vai ser R$ %.2f.", valor_hora*quant_horas))
```

Crie um algoritmo em R que peça a temperatura em graus Farenheit, transforme e mostre a temperatura em graus Celsius.
C = (5 * (F-32) / 9).

```
temp_F=as.numeric(readline("Digite a temperatura em graus Farenheit: "))
print(sprintf("%.2f°F equivale a %.2f°C",temp_F,5*(temp_F-32)/9))
```

Crie um algoritmo em R que peça a temperatura em graus Celsius, transforme e mostre em graus Farenheit.

```
temp_C=as.numeric(readline("Digite a temperatura em graus Celsius: "))
print(sprintf("%.2f°C equivale a %.2f°F",temp_C,9*temp_C/5+32))
```
